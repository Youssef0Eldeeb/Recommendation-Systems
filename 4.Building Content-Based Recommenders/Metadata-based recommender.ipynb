```python
import pandas as pd
import numpy as np

#Import the original file
orig_df = pd.read_csv('/Users/youssefeldeeb/Documents/Material/Recommendation_Systems_with_Python/Datasets/movies_metadata.csv',low_memory=False)

#Import data from the clean file
df = pd.read_csv('/Users/youssefeldeeb/Documents/Material/Recommendation_Systems_with_Python/Datasets/metadata_clean.csv',low_memory=False)

cred_df = pd.read_csv('/Users/youssefeldeeb/Documents/Material/Recommendation_Systems_with_Python/Datasets/credits.csv',low_memory=False)
key_df = pd.read_csv('/Users/youssefeldeeb/Documents/Material/Recommendation_Systems_with_Python/Datasets/keywords.csv',low_memory=False)

# df.head()
```

```python
#Add the useful features into the cleaned dataframe
df['id'] = orig_df['id']
```

```python
# Function to convert all non-integer IDs to NaN
def cleanId(x):
    try:
        return int(x)
    except:
        return np.nan

#Clean the ids 
df['id'] = df['id'].apply(cleanId)
key_df['id'] = key_df['id'].apply(cleanId)
cred_df['id'] = cred_df['id'].apply(cleanId)

#Filter all rows that have a null ID
df = df[df['id'].notnull()]
key_df = key_df[key_df['id'].notnull()]
cred_df = cred_df[cred_df['id'].notnull()]
```

```python
# Convert IDs into integer
df['id'].astype('int')
key_df['id'].astype('int')
cred_df['id'].astype('int')
```

```python
# Merge keywords and credits into your main metadata dataframe
df = df.merge(key_df, on='id')
df = df.merge(cred_df, on='id')
df.head()
```

```python
# Convert the stringified objects into the native python objects
from ast import literal_eval
features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    df[feature] = df[feature].apply(literal_eval)
```

```python
# Extract the director's name. If director is not listed, return NaN
def getDirector(x):
    for crew_num in x:
        if crew_num['job'] == 'Director':
            return crew_num['name']
    return np.nan
```

```python
# Define the new director feature
df['director'] = df['crew'].apply(getDirector)
df.head()
```

```python
# Returns the list top 3 elements or entire list; whichever is more.
def generateList(x):
    if isinstance(x,list):
        names = [element['name'] for element in x]
        #Check if more than 3 elements exist. If yes, return only first three.
        #If no, return entire list.
        if len(names) > 3:
            names = names[:3]
        return names
    #Return empty list in case of missing/malformed data
    return []
```

```python
#Apply the generate_list function to cast and keywords
df['cast'] = df['cast'].apply(generateList)
df['keywords'] = df['keywords'].apply(generateList)

#Only consider a maximum of 3 genres
df['genres'] = df['genres'].apply(lambda x: x[:3])
```

```python
df.head()
```

```python
# Function to sanitize data to prevent ambiguity.
# Removes spaces and converts to lowercase
def sanitize(x):
    if isinstance(x,list):
        #Strip spaces and convert to lowercase
        return [str.lower(i.replace(" ","")) for i in x]
    else:
        #Check if director exists. If not, return empty string
        if isinstance(x,str):
            return str.lower(x.replace(" ",""))
        else:
            return ""
```

```python
#Apply the sanitize function to cast, keywords, director and genres
for feature in ['cast', 'director', 'genres', 'keywords']:
    df[feature] = df[feature].apply(sanitize)
```

```python
df.head()
```

```python
#Function that creates a soup out of the desired metadata
def create_soup(x):
    return ' '.join(x['keywords'])+' '+' '.join(x['cast'])+' '+x['director']+' '+' '.join(x['genres'])
```

```python
# Create the new soup feature
df['soup'] = df.apply(create_soup, axis=1)
```

```python
#Display the soup of the first movie
df.iloc[0]['soup']
```

```python
#Define a new CountVectorizer object and create vectors for the soup
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df['soup'])
```

```python
# Reset index of your df and construct reverse mapping again
df = df.reset_index()
indices2 = pd.Series(df.index, index=df['title'])
indices2.head(10)
```

```python
# Import linear_kernel to compute the dot product
from sklearn.metrics.pairwise import cosine_similarity

# Compute the cosine similarity matrix
cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
cosine_sim2[0:6]
```

```python
# Function that takes in movie title as input and gives recommendations
def contentRecommender(title, cosine_sim, df, indices):
    # Obtain the index of the movie that matches the title
    idx = indices[title]
    
    # Get the pairwsie similarity scores of all movies with that movie
    # And convert it into a list of tuples 
    simScores = list(enumerate(cosine_sim[idx])) 
    
    # Sort the movies based on the cosine similarity scores
    simScores = sorted(simScores, key=lambda x: x[1], reverse=True)
    
    # Get the scores of the 10 most similar movies. Ignore the first movie.
    simScores = simScores[1:11]
    
    # Get the movie indices
    moviesIndices = [i[0] for i in simScores]
    
    # Return the top 10 most similar movies
    return df['title'].iloc[moviesIndices]
    
```

```python
#Get recommendations for The Lion King
contentRecommender('The Lion King', cosine_sim2, df, indices2)
```
